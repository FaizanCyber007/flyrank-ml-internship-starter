## Success Metric

The model is intended to prioritize pages for review rather than produce perfectly calibrated probabilities.

Suitable evaluation metrics include:

- Precision@K
- Recall@K
- NDCG (Normalized Discounted Cumulative Gain)

For this task, Precision@K is especially useful because the content team only reviews a limited number of pages each week. The model is successful if the highest-ranked pages are genuinely at risk of declining.

## Why ML beats a fixed rule

A simple rule such as

"If impressions dropped by 20%, refresh the page"

ignores the interaction between multiple signals.

Some pages naturally fluctuate due to seasonality, while others consistently recover without intervention.

Machine learning can learn complex relationships between historical traffic, impressions, CTR, engagement, and momentum, allowing it to identify subtle warning signs that a fixed threshold would miss.

This results in a better prioritization of pages before traffic loss becomes severe.

## 3. Quick look at the data (2-3 real numbers)

To justify building a predictive model, we need to see the baseline scale of the problem. If a massive volume of high-impact pages is already showing decline patterns, manual review is completely insufficient, making a ranked machine learning queue necessary.

```python
import pandas as pd
import os

!git clone https://github.com/FaizanCyber007/flyrank-ml-internship-starter.git

current_dir = os.getcwd()

repo_root = os.path.abspath(os.path.join(current_dir, '../../'))

csv_path = os.path.join(repo_root, 'data', 'raw', 'content_refresh_anonymized.csv')

if not os.path.exists(csv_path):
    print(f"Error: File not found at {csv_path}")
    data_raw_dir = os.path.join(repo_root, 'data', 'raw')
    if os.path.exists(data_raw_dir):
        print(f"Contents of {data_raw_dir}: {os.listdir(data_raw_dir)}")
    else:
        print(f"Directory {data_raw_dir} does not exist.")
    raise FileNotFoundError(f"CSV file not found: {csv_path}")

df = pd.read_csv(csv_path)

total_pages = len(df)

declining_pages = len(df[df['trend_direction'] == 'down'])
pct_declining = (declining_pages / total_pages) * 100

high_impact_declining = len(df[(df['trend_direction'] == 'down') & (df['impressions_90d'] >= 1000)])

print(f"Total pages in starter dataset: {total_pages}")
print(f"Pages currently showing a decline: {declining_pages} ({pct_declining:.1f}%)")
print(f"High-impact declining pages (1k+ impressions): {high_impact_declining}")
```
